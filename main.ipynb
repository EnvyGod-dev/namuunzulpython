```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score

# huwaaj data train 70 %  test 30 %
x_train, x_test, y_train, y_test = train_test_split(x_1, y, test_size=0.3, random_state=42)

#random forest classifier  n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=43)      
clr_rf = clf_rf.fit(x_train,y_train)

ac = accuracy_score(y_test,clf_rf.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_rf.predict(x_test))
sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# Логистик регресс
model_lr = LogisticRegression(random_state=42).fit(x_train, y_train)
pred_lr = model_lr.predict(x_test)
acc_lr = accuracy_score(y_test, pred_lr)
f1_lr = f1_score(y_test, pred_lr, average='weighted')

# SVM
model_svm = SVC(random_state=42).fit(x_train, y_train)
pred_svm = model_svm.predict(x_test)
acc_svm = accuracy_score(y_test, pred_svm)
f1_svm = f1_score(y_test, pred_svm, average='weighted')

# Random Forest
model_rf = RandomForestClassifier(random_state=42).fit(x_train, y_train)
pred_rf = model_rf.predict(x_test)
acc_rf = accuracy_score(y_test, pred_rf)
f1_rf = f1_score(y_test, pred_rf, average='weighted')

# Үр дүнгүүдийг Харьцуулах
results = {
    "Model": ["Logistic Regression", "SVM", "Random Forest"],
    "Accuracy": [acc_lr, acc_svm, acc_rf],
    "F1 Score": [f1_lr, f1_svm, f1_rf]
}

import pandas as pd
results_df = pd.DataFrame(results)
print(results_df)
```
